```python
def get_max_similarity_per_p(similarities):
    res = []
    for index, value in enumerate(similarities):
        max_indices = thesisUtils.get_n_indexes_of_max_values(value, 6)
        max_indices_without_self = max_indices[:-1]
        max_similarity = value[max_indices_without_self[-1]]
        res.append(max_similarity)
    return res
```

```python
zwickau_tfidf_max_similarity_per_p = get_max_similarity_per_p(zwickau_tfidf_similarities)
london_tfid_max_similarity_per_p = get_max_similarity_per_p(london_tfidf_similarities)
```

```python
zwickau_cv_max_similarity_per_p = get_max_similarity_per_p(zwickau_cv_similarities)
london_cv_max_similarity_per_p = get_max_similarity_per_p(london_cv_similarities)
```

```python
%matplotlib inline
fig, ax = plt.subplots(figsize=(30, 5))
ax.plot(zwickau_tfidf_max_similarity_per_p, label="zwickau tfidf")
ax.plot(zwickau_cv_max_similarity_per_p, label="zwickau cv")
ax.legend()
plt.title('Inner max similarity per p')
plt.show()
```

```python
%matplotlib inline
fig, ax = plt.subplots(figsize=(30, 5))
ax.plot(london_tfid_max_similarity_per_p, label="london tfidf")
ax.plot(london_cv_max_similarity_per_p, label="london cv")
ax.legend()
plt.title('Inner max similarity per p')
plt.show()
```

```python
def cross_version_similarity(version_1_corpus, version_2_corpus, get_features):
    res = []
    for i, p in enumerate(version_1_corpus):
        temp_corpus = [p] + version_2_corpus
        df_features = get_features(temp_corpus)
        temp_similarities = cosine_similarity(df_features, df_features)
        res.append(temp_similarities[0])
    return res
```

```python
zwickau_london_tfidf_similarities = cross_version_similarity(
    zwickau_corpus, 
    london_corpus,
    thesisTfIdfNgramFeatures.create_5_gram
)
london_zwickau_tfidf_similarities = cross_version_similarity(
    london_corpus, 
    zwickau_corpus,
    thesisTfIdfNgramFeatures.create_5_gram
)
```

```python
zwickau_london_cv_similarities = cross_version_similarity(
    zwickau_corpus, 
    london_corpus,
    thesisCountVectorizerNgramFeatures.create_5_gram
)
london_zwickau_cv_similarities = cross_version_similarity(
    london_corpus, 
    zwickau_corpus,
    thesisCountVectorizerNgramFeatures.create_5_gram
)
```

```python
zwickau_london_tfidf_max_similarity_per_p = get_max_similarity_per_p(zwickau_london_tfidf_similarities)
london_zwickau_tfidf_max_similarity_per_p = get_max_similarity_per_p(london_zwickau_tfidf_similarities)
```

```python
zwickau_london_cv_max_similarity_per_p = get_max_similarity_per_p(zwickau_london_cv_similarities)
london_zwickau_cv_max_similarity_per_p = get_max_similarity_per_p(london_zwickau_cv_similarities)
```

```python
fig, ax = plt.subplots(figsize=(30, 5))
ax.plot(zwickau_london_tfidf_max_similarity_per_p, label="zwickau tfidf")
ax.plot(zwickau_london_cv_max_similarity_per_p, label="zwickau cv")
ax.legend()
plt.title('Cross max similarity per p')
plt.show()
```
